```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
diabetes=pd.read_csv("diabetes_model.csv")
```

```python
sns.countplot(diabetes['Outcome'])
```

```python
import seaborn as sns
sns.boxplot(x='Outcome',y='Age',data=diabetes)
```

```python
slice_data=diabetes[['Outcome','Pregnancies','Glucose','BloodPressure','SkinThickness','Insulin','BMI','DiabetesPedigreeFunction','Age'
]]
```

```python
plt.xticks(rotation=45)
plt.figure(figsize=(10,10))
sns.heatmap(slice_data.corr(),annot=True)
```

```python
sns.violinplot(x='Outcome',y='DiabetesPedigreeFunction',data=diabetes)
```

```python
plt.figure(figsize=(8,8))
sns.scatterplot(x='Glucose',y='Insulin',data=diabetes,hue='Outcome')
```
